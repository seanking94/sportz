```python
dat = pd.read_csv('./Shot2017-861/PGATour/Exports/rshot.txt', delimiter = ';', encoding='latin-1', low_memory = False)
```

```python
dat[[' Tournament Name', ' Player Last Name', ' Round', ' Hole', ' Strokes Gained/Baseline', ' Strokes Gained Category']]
```

```python
dat['Course Name'].unique()
```

```python
player_ids = dat[' Player #'].unique()
n_players = len(player_ids)
stroke_gain_mat = np.zeros((n_players, 4))
stroke_types = dat[' Strokes Gained Category'].unique()
```

```python
n_players
```

```python
for i in range(0,n_players):
    dat_player = dat[dat[' Player #'] == player_ids[i]]
    for j in range(0,4):
        dat_player_stroke = dat_player[dat_player[' Strokes Gained Category'] == stroke_types[j]]
        stroke_gain_mat[i,j] = dat_player_stroke[' Strokes Gained/Baseline'].mean()
```
